```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy.signal import find_peaks

# ==========================================
# 1. 関数定義パート
# ==========================================

def get_positive_equilibrium(params):
    """正の平衡点を計算する"""
    r_x1, r_y1, r_x2, r_y2 = params['r']
    l11, l12 = params['lambda'][0]
    l21, l22 = params['lambda'][1]
    c1, c2 = params['c']
    d1, d2 = params['d']

    # 1. 捕食者 y (Preyの式=0より)
    A_y = np.array([[l11, l12], [l21, l22]])
    b_y = np.array([r_x1, r_x2])

    # 2. 被食者 x (Predatorの式=0より)
    A_x = np.array([[c1*l11, d1*l21], [c2*l12, d2*l22]])
    b_x = np.array([r_y1, r_y2])

    try:
        y_eq = np.linalg.solve(A_y, b_y)
        x_eq = np.linalg.solve(A_x, b_x)
        final_z = np.array([x_eq[0], y_eq[0], x_eq[1], y_eq[1]])
        if np.all(final_z > 0): return final_z
        else: return None
    except np.linalg.LinAlgError:
        return None

def model(z, t, params):
    """微分方程式"""
    x1, y1, x2, y2 = z
    r_x1, r_y1, r_x2, r_y2 = params['r']
    l11, l12 = params['lambda'][0]
    l21, l22 = params['lambda'][1]
    c1, c2 = params['c']
    d1, d2 = params['d']
    
    dx1dt = r_x1*x1 - l11*x1*y1 - l12*x1*y2
    dy1dt = -r_y1*y1 + c1*l11*x1*y1 + d1*l21*x2*y1
    dx2dt = r_x2*x2 - l21*x2*y1 - l22*x2*y2
    dy2dt = -r_y2*y2 + c2*l12*x1*y2 + d2*l22*x2*y2
    
    return [dx1dt, dy1dt, dx2dt, dy2dt]

def analyze_periodicity(t, data, name="x1"):
    """
    時系列データの後半を解析し、周期性・収束・発散を判定する
    (numpyだけで動作するように改良)
    """
    # 後半50%のデータを使用
    start_idx = len(data) // 2
    y_steady = data[start_idx:]
    t_steady = t[start_idx:]
    
    # ピーク検出
    peaks, _ = find_peaks(y_steady, distance=50)
    
    # --- 判定ロジック ---
    if len(peaks) < 2:
        # ピークがほとんどない = 直線に近い = 安定平衡点に収束済み
        return "収束 / 定常状態 (Stable Equilibrium)", None, None

    peak_values = y_steady[peaks]
    peak_times = t_steady[peaks]
    
    # 振幅のばらつき (変動係数)
    cv_height = np.std(peak_values) / (np.mean(peak_values) + 1e-9)
    
    # トレンド（傾き）を計算 (numpy.polyfitを使用)
    # 1次関数 (y = ax + b) で近似し、a (傾き) を見る
    slope, intercept = np.polyfit(peak_times, peak_values, 1)

    # 判定
    threshold_stable = 0.005 # 0.5%以下の変動なら安定とみなす
    
    status = ""
    if cv_height < threshold_stable and abs(slope) < 1e-4:
        status = "★ 周期解 (Limit Cycle / Periodic)"
    elif slope < -1e-5:
        status = "減衰振動 (Damped Oscillation -> Converging)"
    elif slope > 1e-5:
        status = "発散振動 (Unstable / Growing)"
    else:
        status = "複雑な挙動 (Complex)"

    print(f"--- {name} の解析結果 ---")
    print(f"状態: {status}")
    print(f"  - 平均振幅: {np.mean(peak_values):.4f}")
    print(f"  - 変動係数: {cv_height:.5f}")
    print(f"  - 傾き: {slope:.6f}")
    
    return status, peak_times, peak_values

# ==========================================
# 2. メイン実行パート
# ==========================================

# パラメータ設定
params = {
    'r': [1.5, 1.0, 1.0, 1.0],

    'lambda': [[0.5, 0.3],

               [0.3, 0.5]],

    'c': [1.01, 0.8],
    'd': [0.8, 1.01]

}

# 平衡点の計算
eq_point = get_positive_equilibrium(params)

if eq_point is None:
    print("エラー: 正の平衡点が見つかりません。")
    base_z0 = np.array([0.5, 0.5, 0.5, 0.5])
else:
    print(f"計算された平衡点: {eq_point}")
    base_z0 = eq_point

# ★初期値の調整 (ここを0.0にすると完全平衡、値をいれると振動開始)
diff_x1 = 2.0   # x1 を少し増やす
diff_y1 = 2.0
diff_x2 = 1.5
diff_y2 = 1.5

z0 =np.array([diff_x1, diff_y1, diff_x2, diff_y2])
z0 = np.maximum(z0, 0) # 負の値防止

# シミュレーション実行
t = np.linspace(0, 500, 100000) # 時間を少し長く、細かくしました
z = odeint(model, z0, t, args=(params,))
x1, y1, x2, y2 = z[:, 0], z[:, 1], z[:, 2], z[:, 3]

# ==========================================
# 3. 解析と可視化パート (リターンマップ付き)
# ==========================================


# 解析実行 (x1について)
verdict, pt, pv = analyze_periodicity(t, x1, name="x1")

plt.figure(figsize=(15, 8)) # 横幅を少し広げました

# --- 図1: 時系列 (左上) ---
plt.subplot(2, 3, 1)
plt.plot(t, x1, label='$x_1$')
plt.plot(t, y1, label='$y_1$', alpha=0.7)
plt.title(f'Time Series\nResult', fontsize=10)
plt.legend()
plt.grid(True)

# --- 図2: 解析結果・ピーク検出 (中上) ---
plt.subplot(2, 3, 2)
start_idx = len(t) // 2
plt.plot(t[start_idx:], x1[start_idx:], label='Last 50% Waveform', color='green', alpha=0.5)
if pt is not None:
    plt.scatter(pt, pv, color='red', marker='x', s=50, label='Peaks', zorder=5)
plt.title('Peak Detection', fontsize=10)
plt.legend()
plt.grid(True)

# --- ★図3: リターンマップ (右上) ---
plt.subplot(2, 3, 3)
if pv is not None and len(pv) > 2:
    # x_n (現在のピーク) vs x_{n+1} (次のピーク)
    peak_n = pv[:-1]
    peak_next = pv[1:]
    
    plt.scatter(peak_n, peak_next, color='purple', s=40, zorder=3, label='Return Map')
    
    # 補助線: y = x (この線上に集まれば1周期解)
    min_val, max_val = min(peak_n), max(peak_n)
    margin = (max_val - min_val) * 0.1
    plt.plot([min_val-margin, max_val+margin], [min_val-margin, max_val+margin], 
             'k--', alpha=0.3, label='y=x (Identity)')
    
    plt.xlabel('$x_n$ (Peak $n$)')
    plt.ylabel('$x_{n+1}$ (Peak $n+1$)')
    plt.title('Return Map (Poincaré Map)', fontsize=10)
    plt.legend()
else:
    plt.text(0.5, 0.5, "No peaks detected\n(Stable or Converged)", 
             ha='center', va='center')
    plt.title('Return Map')
plt.grid(True)

# --- 図4: 位相平面 x1 vs y1 (左下) ---
plt.subplot(2, 3, 4)
plt.plot(x1, y1, color='tab:blue', alpha=0.8)
plt.plot(base_z0[0], base_z0[1], 'ko', label='Eq Point') # 平衡点
plt.plot(z0[0], z0[1], 'go', label='Start')              # スタート地点
plt.xlabel('$x_1$'); plt.ylabel('$y_1$')
plt.title('Phase Portrait ($x_1$ vs $y_1$)', fontsize=10)
plt.legend()
plt.grid(True)

# --- 図5: 位相平面 x2 vs y2 (中下) ---
plt.subplot(2, 3, 5)
plt.plot(x2, y2, color='tab:orange', alpha=0.8)
plt.xlabel('$x_2$'); plt.ylabel('$y_2$')
plt.title('Phase Portrait ($x_2$ vs $y_2$)', fontsize=10)
plt.grid(True)

# --- 図6: 3D軌道図 (右下・おまけ) ---
ax = plt.subplot(2, 3, 6, projection='3d')
ax.plot(x1, x2, y1, alpha=0.7, color='teal')
ax.set_xlabel('$x_1$')
ax.set_ylabel('$x_2$')
ax.set_zlabel('$y_1$')
ax.set_title('3D Trajectory ($x_1, x_2, y_1$)', fontsize=10)

plt.tight_layout()
plt.savefig("c1_d2_d_6cycle")
plt.show()
```
